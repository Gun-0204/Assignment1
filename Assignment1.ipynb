Question 1.Explain the main components of client-server architecture and their functions.

Answer ->  Client-Server architecture is a model where client communicate with servers to request and receive data over a network.

Main components include:

1. Client:

  .A client is a device that request services from a server.

  .Example-when you open a website, your browser requests the web page from the server.

2. Server:

  .A server is a powerful computer that:

   1. collects request from client.

   2. process the request.

   3. create a response and send it back to the client.

  .Example-A web browser stores wesites and sends them when requested.

3. Network Protocol:

  .Network protocol is a set of rules that define how data is transmitted between clients and servers.

  .Examples-.HTTPS/HTTPS,SMTP,FTP etc.

4. IP Address:

  .A unique identifier for device on a network.

  .Example-192.168.1.1

5. Port:

  .A virtual communication endpoint on a device used for different type of services.

  .Example-Port 80, Port 443 etc.

6. Socket Address:

  .A combination of IP addresses+Port number to specify where to send/receive data.

  .Example-192.168.1.1:443

Question 3. Describe the role of a web browser's developer tools in web development.

Answer->

Web browser's Developer Tools (DevTools) help web developers test, debug, and improve websites easily. They are built into browsers like Chrome, Firefox, and Edge.

Key Roles of DevTools in Web Development:

1. Inspect and Edit Web Pages

  .Allows developers to view and edit HTML, CSS, and JavaScript in real time.

  .Helps fix layout issues without changing the actual code file.

  .Example: You can change the font size or color of a webpage instantly.


2. Debug JavaScript Errors

  .The Console tab shows errors and warnings in JavaScript.

  .Developers can test and run JavaScript commands directly.

  .Example: If a button isn’t working, you can check for errors in the console.


3. Analyze Website Speed and Performance

  .The Network tab tracks loading times and finds slow resources.

  .Helps developers optimize images, scripts, and CSS for faster websites.

  .Example: If a website is slow, you can see which file is causing the delay.


4. Responsive Design Testing

  .The Device Toolbar lets developers test how websites look on mobile, tablet, and desktop screens.

  .Ensures a website works well on different devices.

  .Example: You can check how a webpage looks on an iPhone or Android without needing a physical device.


5. Manage Cookies and Storage

  .The Application tab shows cookies, local storage, and session storage.

  .Helps developers manage user sessions and store data securely.

  .Example: You can delete cookies to see how a website behaves for a first-time user.


6. Monitor API Calls

  .The Network tab helps track API requests between the frontend and backend.

  .Developers can check if data is sent and received correctly.

  .Example: If a form submission isn’t working, you can see if the request was sent to the server.


 Question 4. What are the advantage of using a Version Control System like Git in a development project ?

Answer-> Git is a Version Control System(VCS) that help developers track changes in their code,collaborate easily, and avoid losing important work.

Here are the key benefits of using Git in a development project:

1.Tracks code changes-It keeeps a history of all code changes, allowing developers to track modification over time.

2.Facilitates Team Collaboration-Multiple developers can work on the same project without overwriting each other's work.

3.Open source and free-Git is completely free and works on multiple platforms.

4.Work across different computers-Developer can clone repositories and work from different machines.



Question 5.Compare and contrast a text editor and an IDE,highlighting their key features and uses.


Answer->  A text editor and IDE(Integrated Development Environment) are tools used by the developers for writing and editing the code.

Text Editor- A text editor is a software used for creating ,editing plain text files,including the code file also.

Some features are listed below:

1. Syntax Highlighting-Color-coded highlighting of code elements of better readability.

2. Code Completion-Suggestions and auto-completion of code snippets,functions and variables.

3. Find and replace-Search for speific text and replace it with desired content.

4. Customization-Ability to customize themes,preferences, and plugins to personalize the editing environments.

5. Light-Weight-Text editors are generally lightweight,fast and consume fewer system resources.

6. Basic Debugging-Limited or no built-in debugging capabilities.

IDE(Integrated Development Environment)- is a software suite that combines a text editor with additional tools and features to facilitate software development.

Some features are listed below:

1. Advanced editing capabilities-Similar to text editor,IDE's after syntax highlighting,code completion, and code formatting features.

2. Integrated compiler and debugger-Built in tools for compiling and debugging code, allowing developers to find and fix errors directly within the IDE.

3. Integrated testing-Tools for unit testing,automated testing and test result analysis.

4. Profiling and performance analysis-Capabilities to profile code execution and analysis performance bottlenecks.

5. Language specific tools-IDE's often include language specific features such as refactoring tools,code templates, and documentation integration.
